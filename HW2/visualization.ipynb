```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read sales.csv data and do the following exercises:


```python
df = pd.read_csv('sales.csv')
df.head()
```

Exercise 1:<br/> Read Total profit of all months and show it using a line plot
Total profit data provided for each month.<br/>
Generated line plot must include the following properties:

X label name = Month Number<br/>
Y label name = Total profit<br/>
The line plot graph should look like this.<br/>




```python
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'])
plt.xticks(ticks=range(1, 13))
plt.yticks(ticks=range(100000, 600000, 100000))
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.show()
```

Exercise 2:<br/>
Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

Line Style dotted and Line-color should be red<br/>
Show legend at the lower right location.<br/>
X label name = Month Number<br/>
Y label name = Sold units number<br/>
Add a circle marker.<br/>
Line marker color as read<br/>
Line width should be 3<br/>
The line plot graph should look like this.<br/>




```python
plt.figure()
plt.plot(df['month_number'], df['total_profit'], linestyle='--', color='red', 
         marker='o', markerfacecolor='black', linewidth=3, label='Profit data of last year')
plt.xticks(ticks=range(1, 13))
plt.yticks(ticks=range(100000, 600000, 100000))
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.title('Company sales data of last year')
plt.show()
```

Exercise 3:<br/>
Read all product sales data and show it using a multiline plot<br/>
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).<br/>

The graph should look like this.<br/>




```python
plt.figure()
for product in ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']:
    plt.plot(df['month_number'], df[product],marker='o',linewidth=3, label=product)
plt.xticks(ticks=range(1, 13))
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Month number')
plt.legend()
plt.show()
```

Exercise 4:<br/>
Read toothpaste sales data of each month and show it using a scatter plot<br/>
Also, add a grid in the plot. gridline style should “–“.<br/>

The scatter plot should look like this.<br/>




```python
plt.figure()
plt.scatter(df['month_number'], df['toothpaste'], label='Toothpaste Sales data')
plt.xticks(ticks=range(1, 13))
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.title('Toothpaste sales data')
plt.grid(linestyle='--')
plt.legend()
plt.show()
```

Exercise 5:<br/>
Read face cream and facewash product sales data and show it using the bar chart<br/>
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.<br/>

The bar chart should look like this.<br/>




```python
plt.figure()
plt.bar(df['month_number'] - 0.2, df['facecream'], width=0.4, label='Face Cream')
plt.bar(df['month_number'] + 0.2, df['facewash'], width=0.4, label='Face Wash')
plt.xticks(ticks=range(1, 13))
plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')
plt.grid(linestyle='--')
plt.title('Facewash and faceframe sales data')
plt.legend()
plt.show()
```

Exercise 7:<br/>
Read the total profit of each month and show it using the histogram to see the most common profit ranges<br/>
The histogram should look like this.<br/>




```python
plt.figure()
plt.hist(
    df['total_profit'], 
    bins=[150000, 175000, 200000, 225000, 250000, 300000, 350000], 
    alpha=0.7, 
    label='Profit data'
)
plt.xticks([150000, 175000, 200000, 225000, 250000, 300000, 350000])
plt.yticks(ticks=range(0, 6))
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.title('Profit data')
plt.legend(loc='upper left')
plt.show()
```

Exercise 8:<br/>
Calculate total sale data for last year for each product and show it using a Pie chart<br/>
Note: In Pie chart display Number of units sold per year for each product in percentage.<br/>

The Pie chart should look like this.<br/>




```python
total_sales = [df[product].sum() for product in ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]
products = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
labels = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=products, autopct='%1.1f%%', startangle=140)
plt.title('Sales data')
plt.legend(labels, loc='best')
plt.show()
```

Exercise 9:<br/>
Read Bathing soap facewash of all months and display it using the Subplot<br/>
The Subplot should look like this.<br/>




```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap', color='black', marker='o')
plt.xticks(ticks=range(1, 13))
plt.yticks([7500,10000,12500])
plt.xlabel('Month Number')
plt.title("Sales data of Bathingsoap")

plt.subplot(2, 1, 2)
plt.plot(df['month_number'], df['facewash'], label='Face Wash', color='red', marker='o')
plt.xticks(ticks=range(1, 13))
plt.yticks([1500,2000])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title("Sales data of facewash")

plt.tight_layout()
plt.show()
```

Exercise 10:<br/>
Read all product sales data and show it using the stack plot<br/>
The Stack plot should look like this.<br/>




```python
plt.figure()
plt.stackplot(df['month_number'], df['facecream'], df['facewash'], df['toothpaste'], 
              df['bathingsoap'], df['shampoo'], df['moisturizer'], 
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Monthly Sales of All Products')
plt.legend(loc='upper left')
plt.show()
```
